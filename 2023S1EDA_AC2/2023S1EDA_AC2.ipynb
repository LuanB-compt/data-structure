## Exercício 1 

Escreva uma função chamada **compara** que receba como parâmetro uma pilha e uma fila. Ela deve retornar True caso as duas estruturas passadas como parâmetro sejam iguais e False caso contrário. Um pilha e uma fila são consideradas iguais tanto quando os elementos estão na mesma ordem ou em ordem contrária. Por exemplo, a pilha A=[0,1,2,3,4] é igual as filas B = [0,1,2,3,4] e C = [4,3,2,1,0].

 - Você deve usar as classes Pilha e Fila ensinadas na disciplina e disponibilizadas juntamente com este notebook. 

- As únicas operações permitidas para acessar os valores da pilha e fila são, respectivamente, as operações pop e dequeue. 

- Você deve garantir que ao final da função, a pilha e fila passadas como parâmetro estejam intactas, isto é, como os mesmos valores passados como entrada. 

````python
################# ESCREVA SEU CÓDIGO AQUI  ###############################

from utils.pilhaDinamica import Pilha
from utils.filaDinamica import Fila


def compara(stack: Pilha, queue: Fila) -> bool or None:
    """
        Check a stack and a queue, if they have the same values in the same order
        or in reverse order, true will be returned. Else, will be returned false.
        If the stack and the queue have no values, will be return None.

        This function has an asymptotic complexity of: O(n) - Linear.

        Parameters: 
            stack (Pilha): a dynamic stack
            queue (Fila): a dynamic queue

        Returns:
            Boolean or None
    """
    # If have no values in stack and queue, will return None
    if stack.tamanho == 0 and queue.size == 0:
        return None
    # If the data structure have different size, it can't be equals
    elif stack.tamanho != queue.size:
        return False
    
    # Creating lists with the values from the stack and queue
    stack_values, queue_values = [], []
    # --- we know that the stack and the queue have the same size
    # --- so it possible use the same loop to create the lists
    while stack.tamanho > 0: 
        stack_values.append(stack.pop())
        queue_values.append(queue.dequeue())

    # Reconstructing the stack and the queue with the lists
    for value in stack_values:
        stack.push(value)
    for value in queue_values:
        queue.enqueue(value)

    # Checking if the the values is in the same order
    for stack_counter in enumerate(stack_values):
        if stack_counter[1] == queue_values[stack_counter[0]]:
            continue
        else:
            break
    else: # If nothing happen inside the loop, will be return True
        return True

    # Checking if the values is in inverse order
    queue_counter = len(queue_values) - 1
    for stack_counter in enumerate(stack_values):
        if stack_counter[1] == queue_values[queue_counter]:
            queue_counter -= 1
            continue
        else:
            return False
    else: # If nothing happen inside the loop, will be return True
        return True

##########################################################################

# testando a função criada
pilha1 = Pilha()
pilha1.push(0)
pilha1.push(1)
pilha1.push(2)
pilha1.push(3)
pilha1.push(4)

# testando a função criada
pilha2 = Pilha()
pilha2.push(4)
pilha2.push(3)
pilha2.push(2)
pilha2.push(1)
pilha2.push(0)

fila1 = Fila()
fila1.enqueue(0)
fila1.enqueue(1)
fila1.enqueue(2)
fila1.enqueue(3)
fila1.enqueue(4)

fila2 = Fila()
fila2.enqueue(4)
fila2.enqueue(3)
fila2.enqueue(2)
fila2.enqueue(1)
fila2.enqueue(0)

fila3 = Fila()
fila3.enqueue(4)
fila3.enqueue(3)
fila3.enqueue(2)
fila3.enqueue(0)
fila3.enqueue(1)

comp1 = compara(pilha1, fila1)
comp2 = compara(pilha1, fila2)
comp3 = compara(pilha2, fila1)
comp4 = compara(pilha2, fila2)
comp5 = compara(pilha2, fila3)

print('Resultado retornado:')
print('\tComparacao 1?: ', comp1)
print('\tComparacao 2?: ', comp2)
print('\tComparacao 3?: ', comp3)
print('\tComparacao 4?: ', comp4)
print('\tComparacao 5?: ', comp5)

print('\n' + 20*'-')
print('Resultado esperado:')
print('\tComparacao 1?: True')
print('\tComparacao 2?: True')
print('\tComparacao 3?: True')
print('\tComparacao 4?: True')
print('\tComparacao 5?: False')
````

## Exercício 2 

Faça uma classe chamada **ListaNSE**. Essa classe deve implementar uma **lista não sequencial e estática**. Essa lista deve ter o mesmo comportamento de uma **lista não sequencial e dinâmica**, mas deve ser guardada dentro de uma lista em Python com capacidade limitada.

Essa classe deve possui os seguintes métodos:

- inserir(x, i): insere x na posição i da estrutura
- remover(i): remove o item que está na posição i da estrutura e retorna seu valor 
- buscar(x): buscar o item x na lista e retorna sua posição
- size(): retorna o número de elementos da lista
- clear(): remove os elementos da lista
- \_\_str\_\_(): imprime a estrutura

- Você deve usar a lista do Python apenas para inicializar a estrutura. Depois disso, nenhum método do Python pode ser usado, como append, insert, etc.

````python
################# ESCREVA SEU CÓDIGO AQUI  ###############################

class ListaNSE:
  """
    Implementation of a non-sequential static list.
  """

  def __init__(self, capacity: int):
    """
      Constructor method.

      Args:
          capacity (int): the list capacity.
    """
    self.data = [None] * capacity
    self.capacity = capacity
    self.amount = 0

  def insert(self, value: any, position: int) -> None:
    """
      Adds the value at the defined position.

      Args:
          value (any): value that will be inserted.
          position (int): the position in list that will be inserted the value.
    """
    if (self.data[position] != None):
      for i in range(self.amount-1, position-1, -1):
        self.data[i+1] = self.data[i]
    self.data[position] = value
    self.amount += 1

  def remove(self, position: int) -> None:
    """
      Sets the value to null at the defined position.

      Args:
          position (int): the position in list that will be removed a value.
    """
    if (self.data[position] != None):
      self.data[position] = None
    self.amount -= 1

  def search(self, value: any) -> bool:
    """
      Linear search in list.

      Args:
          value (any): the element that will be searched in list.

      Returns:
          bool: if found the element, will be returned True, else will be returned False.
    """
    for dado in self.data:
      if dado == value:
        return True
    return False

  def size(self) -> int:
    """
      Get the number os elements in the list.

      Returns:
          int: number of elements in the list.
    """
    return self.amount

  def clear(self):
    self.data = []
    self.amount = 0

  # Str function: 
  def __str__(self) -> str:
    """
      Prints only non-null value from the non-sequential static list.

      Returns:
          str: string that will be printed.
    """
    text = "["
    for item in self.data:
      if item is not None:
        text += str(item) + ","
    if len(self.data) > 0:
        text = text[:-1]
    text += "]"
    return text

##########################################################################

# testa a Lista
lista = ListaNSE(5)        

# inserir no inicio da lista
lista.insert(5,0)
print( 'insert(5,0): ', lista )

lista.insert(10,0)
print( 'insert(10,0): ', lista )

lista.insert(24,2)
print( 'insert(24,2): ', lista )

lista.insert(25,1)
print( 'insert(25,1): ', lista )

lista.remove(2)
print( 'remove(2): ', lista )

print( 'Size: ', lista.size() )

lista.clear()
print( 'Clear: ', lista )

print('\n' + 20*'-')
print('Resultado esperado:')
print( 'insertHead(5,0): [5]')
print( 'insertHead(10,0): [10,5]')
print( 'insert(24,2): [10,5,24]')
print( 'insert(25,1): [10,25,5,24]')
print( 'remove(2): [10,25,24]')
print( 'Size: 3')
print( 'Clear: []')
````

````python
################# ESCREVA SEU CÓDIGO AQUI  ###############################

from utils.ListaDuplamenteLigada import *


class ListaInsertion(Lista):
    
    def insert(self, valor: any) -> None:
        """
            Method to insert a new value in the list.

            Args:
                valor (any): the value that will be inserted.
        """
        no = Node(valor)
        if self.head is None:
            self.head = no
            self.tail = no
        else:
            atual = self.head
            while atual is not None and self.comparar(valor, atual.data) > 0:
                atual = atual.next

            if atual is None:
                no.prev = self.tail
                self.tail.next = no
                self.tail = no
            else:
                no.next = atual
                no.prev = atual.prev
                if atual.prev is None:
                    self.head = no
                else:
                    atual.prev.next = no
                atual.prev = no
        self.size += 1

    def comparar(self, str1: str, str2: str) -> float:
        """
            Method to compare two strings and return the difference between the
            averages of the strings.

            Args:
                str1 (str): string to compare.
                str2 (str): string to compare.

            Returns:
                float: return the difference between two strings.
        """
        media1 = int(sum(len(palavra) for palavra in str1.split()) / len(str1.split()))
        media2 = int(sum(len(palavra) for palavra in str2.split()) / len(str2.split()))
        return media1 - media2

    def sublista(self, a: float or int, b: float or int) -> Lista:
        """
            Method to make a new list with only the values that are between positions a and b.

            Args:
                a (float or int): value to search the positions.
                b (float or int): value to search the positions.

            Returns:
                Lista: the sublist between the positions a and b.
        """
        if a < 0 or b < 0 or a >= self.size or b >= self.size or a > b:
            return Lista()
        sub = Lista()
        atual = self.head
        i = 0
        while i < a:
            atual = atual.next
            i += 1
        while i <= b:
            sub.inserir(atual.data)
            atual = atual.next
            i += 1
        return sub

    def repeticoes(self, num_media: int or float, inicio: int=0, fim: int or None=None) -> int:
        """
            Takes a number and uses a modified recursive binary search to return the number
            of strings that contain the same average number of characters.

            Args:
                num_media (int or float): number with the average numbers.
                inicio (int, optional): control the end of the recursive binary search. Defaults to 0.
                fim (intorNone, optional): control the end of the recursive binary search. Defaults to None.

            Returns:
                int: _description_
        """
        if fim is None:
            fim = self.size - 1
        if inicio > fim:
            return 0
        meio = (inicio + fim) // 2
        atual = self.head
        for i in range(meio):
            atual = atual.next
        media_meio = int(sum(len(palavra) for palavra in atual.data.split()) / len(atual.data.split()))
        if media_meio == num_media:
            return 1 + self.repeticoes(num_media, inicio, meio - 1) + self.repeticoes(num_media, meio + 1, fim)
        elif media_meio > num_media:
            return self.repeticoes(num_media, inicio, meio - 1)
        else:
            return self.repeticoes(num_media, meio + 1, fim)

##########################################################################

# testa o exercicio
lista = ListaInsertion()        

# inserir no inicio da lista
lista.insert("Centro Universitario Facens")
lista.insert("Eu fui para um shopping em Sorocaba")
lista.insert("A menina gosta de jogar")
lista.insert("O menino gosta de andar")
lista.insert("Tá tá tá")

print("\nLista:\n", lista)

qtd1 = lista.repeticoes(3)
qtd2 = lista.repeticoes(2)
qtd3 = lista.repeticoes(30)
print("Qtd. repeticoes 1: ", qtd1)
print("Qtd. repeticoes 2: ", qtd2)
print("Qtd. repeticoes 3: ", qtd3)

print('\n' + 20*'-')
print('Resultado esperado:')
print("\nLista:\n")
print('[Tá tá tá, A menina gosta de jogar, ', end="")
print('O menino gosta de andar, ', end="")
print('Eu fui para um shopping em Sorocaba, ', end="")
print('Centro Universitario Facens]', end="")
````

## Exercício 5

Crie uma classe chamada **PilhaFila** que implemente uma estrutura de dados que seja ao mesmo tempo uma pilha e fila dinâmica usando um deque dinâmico instanciado a partir da classe fornecida junto com esse notebook. Você deve implementar todos os métodos de uma fila e de um pilha dinâmica nessa estrutura aproveitando tudo o que for possível do deque. 

````python
################# ESCREVA SEU CÓDIGO AQUI  ###############################

from utils.dequeDinamico import Deque
from utils.filaDinamica import Fila
from utils.pilhaDinamica import Pilha


class pilhaFila():
    
    def __init__(self):
        """
            Constructor method.
        """
        self.deque = Deque()
    
    # --- Stack ---
    def push(self, valor: any) -> None:
        """
            Method to push a value in to structure.

            Args:
                valor (any): value to push a value in to structure.
        """
        self.deque.enqueueRear(valor = valor)

    def pop(self) -> any:
        """
            Method to remove the top elemente from the structure.

            Returns:
                any: return the top element from the structure.
        """
        return self.deque.dequeueRear()

    def top(self) -> any:
        """
        Return the top element from the structure, but don't remove this element.

        Returns:
            any: return the top element from the structure.
        """
        return self.deque.getRear()

    # --- Queue ---
    def enqueue(self, valor: any) -> None:
        """
            Method to enqueue a value in to structure.

            Args:
                valor (any): value to enqueue in the structure.
        """
        self.deque.enqueueRear(valor = valor)

    def dequeue(self) -> any:
        """
            Method to remove the first element that joined in the structure.

            Returns:
                any: the first element that joined in the structure.
        """
        return self.deque.dequeueFront()

    def front(self) -> any:
        """
            Method to get the first elemente that joined in the structure, but
            don't remove this element.

            Returns:
                any: the first element that joined in the structure.
        """
        return self.deque.getFront()

    def rear(self) -> any:
        """
            Method to get the last elemente that joined in the structure, but
            don't remove this element.

            Returns:
                any: the last element that joined in the structure.
        """
        return self.deque.getRear()

    # --- Common ---
    def clear(self) -> None:
        """
            Method to clear all the structure.
        """
        self.deque.clear()

    def isEmpty(self) -> bool:
        """
            Method to check if is empty the structure.

            Returns:
                bool: if is empty, will be return True, else will be returned False.
        """
        return self.deque.isEmpty()

    def size(self) -> int:
        """
            Method to get the size of the structure.

            Returns:
                int: the size of the structure.
        """
        return self.deque.getSize()
    
    def __str__(self) -> str:
        """
            Create the string to print the structure.

            Returns:
                str: string that will be printed.
        """
        return self.deque.__str__()
    
##########################################################################
       
# testando a PilhaFila
pilhaFila1 = pilhaFila()

pilhaFila1.push(5)
print(pilhaFila1)

pilhaFila1.enqueue(3)
print(pilhaFila1)

pilhaFila1.push(9)
print(pilhaFila1)

print( 'Size: ', pilhaFila1.size() )

valor = pilhaFila1.pop()
print(pilhaFila1)

valor = pilhaFila1.dequeue()
print(pilhaFila1)

print( 'Vazia?: ', pilhaFila1.isEmpty() )

valor = pilhaFila1.pop()
print(pilhaFila1)

print( 'Vazia?: ', pilhaFila1.isEmpty() )

valor = pilhaFila1.pop()
print(pilhaFila1)

pilhaFila1.push(2)
print(pilhaFila1)

pilhaFila1.push(14)
print(pilhaFila1)

print(pilhaFila1.front())
print(pilhaFila1.rear())
print(pilhaFila1.top())

print( 'Vazia?: ', pilhaFila1.isEmpty() )

pilhaFila1.clear()
print( 'Vazia?: ', pilhaFila1.isEmpty() )

print('\n' + 20*'-')
print('Resultado esperado:')
print("\nLista:\n")
print('[ 5 ]')
print('[ 3, 5 ]')
print('[ 3, 5, 9 ]')
print('Size:  3')
print('[ 3 5 ]')
print('[ 3 ]')
print('Vazia?:  False')
print('[ ]')
print('Vazia?:  True')
print('pilhaFila1 vazia')
print('[ ]')
print('[ 2 ]')
print('[ 2 14 ]')
print('2')
print('14')
print('14')
print('Vazia?:  False')
print('Vazia?:  True)')
````

Essa função deverá verificar se as duas árvores são iguais. Ás árvores são iguais caso elas tenham exatamente a mesma estrutura e os mesmos valores armazenados exatamente nos mesmos nós. 

````python
################# ESCREVA SEU CÓDIGO AQUI  ###############################

from utils.arvore_binaria_de_busca import Tree


def compara(tree1: Tree, tree2: Tree) -> bool:
    """
        Recursive function that receives two trees instantiated from the binary search
        tree class and checks whether the two trees are equal.

        Args:
            tree1 (Tree): tree to compare.
            tree2 (Tree): tree to compare.

        Returns:
            bool: if the trees is equal, will be returned True, else will be returned False.
    """
    #caso as árvores sejam vazias, elas são iguais, então esse caso retorna true
    if tree1 is None and tree2 is None:
        return True
    #caso uma das árvores seja vazia e a outra não, elas são diferentes, então o retorno é false
    elif tree1 is None and tree2 is not None:
        return False
    # caso uma das árvores seja vazia e a outra não, elas são diferentes, então o retorno é false
    elif tree1 is not None and tree2 is None:
        return False
    #caso ambos os nós raíz sejam vazios, então ambas as árvores são iguais nesse ponto, então é retornado true
    if tree1.root is None and tree2.root is None:
        return True
    #caso um dos nós seja vazio e o outro não, as árvores não são iguais, então deve ser retornado false
    elif tree1.root is None and tree2.root is not None:
        return False
    #caso um dos nós seja vazio e o outro não, as árvores não são iguais, então deve ser retornado false
    elif tree1.root is not None and tree2.root is None:
        return False
    #Agora chegamos na verificação dos dados em si, se o valor do nó da primeira árvore não for igual o da segunda árvore, então é retornado false
    if tree1.root.data != tree2.root.data:
        return False

    #neste bloco é definida a subárvore da esquerda da árvore1 que sera utilizada na chamada recursiva da função compara

    arvore_esquerda1 = Tree()
    if tree1.root.left is None:
        arvore_esquerda1.root = None
    else:
        arvore_esquerda1.root = tree1.root.left

    #neste bloco é definida a subárvore da esquerda da árvore2 que sera utilizada na chamada recursiva da função compara

    arvore_esquerda2 = Tree()
    if tree2.root.left is None:
        arvore_esquerda2.root = None
    else:
        arvore_esquerda2.root = tree2.root.left

    #neste bloco é definida a subárvore da direita da árvore1 que sera utilizada na chamada recursiva da função compara
    arvore_direita1 = Tree()
    if tree1.root.right is None:
        arvore_direita1.root = None
    else:
        arvore_direita1.root = tree1.root.right

    #neste bloco é definida a subárvore da direita da árvore2 que sera utilizada na chamada recursiva da função compara

    arvore_direita2 = Tree()
    if tree2.root.right is None:
        arvore_direita2.root = None
    else:
        arvore_direita2.root = tree2.root.right
    resultEsquerda = compara(arvore_esquerda1, arvore_esquerda2)
    resultDireita = compara(arvore_direita1, arvore_direita2)

    return resultEsquerda and resultDireita

##########################################################################

tree1 = Tree()
tree1.insert(6)
tree1.insert(2)
tree1.insert(7)
tree1.insert(1)
tree1.insert(4)
tree1.insert(3)
tree1.insert(5)

tree2 = Tree()
tree1.insert(6)
tree1.insert(2)
tree1.insert(7)
tree1.insert(1)
tree1.insert(3)
tree1.insert(4)
tree1.insert(5)

print(compara(tree1, tree1))
print(compara(tree1, tree2))

print('\n' + 20*'-')
print('Resultado esperado:')
print("False")
print("True")
````

Modifique a árvore binária de busca ensinada nesta disciplina e dê o nome de **TreePath**. Acrescente nessa árvore um método chamado **caminhos**. Esse método deve preencher um objeto da lista duplamente ligada fornecida junto com este notebook com a soma do valores dos nós que estão no caminho de cada um dos nós folhas até a raiz da árvore. Por exemplo, considere a árvore apresentada abaixo:

```
            ___________61___________
           |                        |
       ____34_____              ____67____
      |           |            |          |    
      19__        50__      __66        __94__
          |           |    |           |      |  
         20           55   64          70     95
```

Se a árvore acima fosse apresentada como entrada, o método **caminhos** deveria retornar uma lista duplamente ligada com os seguintes valores: [134, 200, 258, 292, 317], que é formada pelas seguites somas: [61+34+19+20, 61+34+50+55, 61+67+66+64, 61+67+94+70, 61+67+94+95].

````python
################# ESCREVA SEU CÓDIGO AQUI  ###############################

from utils.arvore_binaria_de_busca import Node, Tree
from utils.ListaDuplamenteLigada import Lista


class TreePath(Tree, Lista):

    def __init__(self):
        """
            Constructor method.
        """
        super().__init__()

    def caminhos(self, node: Node = None, amount: int = 0, lista: Lista = None):
        """
            Method to fills a doubly linked list object with the sum of the values of the nodes
            that are on the path from each of the leaf nodes to the root of the tree.

            Args:
                node (Node, optional): node to start the calculation. Defaults to None.
                amount (int, optional): Defaults to 0.
                lista (Lista, optional): the returned list. Defaults to None.

            Returns:
                _type_: _description_
        """
        
        # In this function, recursion is used to traverse the tree starting from the initial node and 
        # accumulate the sum of the values that are in the path from each leaf node to the root of the tree.
        
        if self.root is None:
            return []
        
        if node is None:
            node = self.root
            
        if lista is None:
            lista = Lista()
        
        if node.left is None and node.right is None:
            if node.left:
                lista.insertHead(amount + node.data)
            else:
                lista.insertTail(amount + node.data)
        
        # The function traverses the subtree of the current tree, accumulating the sum of the values of the nodes along the way.

        if node.left:
            self.caminhos(node.left, amount + node.data, lista)
        
        if node.right:
            self.caminhos(node.right, amount + node.data, lista)
        
        return lista


##########################################################################

tree1 = TreePath()
tree1.insert(61)
tree1.insert(34)
tree1.insert(67)
tree1.insert(19)
tree1.insert(50)
tree1.insert(66)
tree1.insert(94)
tree1.insert(20)
tree1.insert(55)
tree1.insert(64)
tree1.insert(70)
tree1.insert(95)

lista = tree1.caminhos()
print(lista)

print('\n' + 20*'-')
print('Resultado esperado:')
print("[134, 200, 258, 292, 317]")
````
